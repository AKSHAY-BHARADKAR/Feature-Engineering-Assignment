```python
df.corr()
```                                                                                                                                                                                                                      
Q. 17. What is causation? Explain Difference between correlation and causation with an example.

Answer :-

- Causation**: One variable causes another.
- Correlation**: Variables move together but don’t necessarily affect each other.
Example : Ice cream sales and drowning incidents correlate, but hot weather causes both.

Q. 18. What is an Optimizer? What are different Types of optimizers? Explain each with an example.

Answer :-

An optimizer is an algorithm used in machine learning and deep learning to adjust the model parameters (like weights and biases) to minimize the loss function.

In simple terms:

The optimizer helps your model learn by tweaking it in the right direction to make better predictions.

The goal is to find the best parameters (usually using gradient descent) so that the loss/error is as low as possible.

Types of Optimizers (with examples)

There are several types of optimizers. Here are the most common ones:

1. Gradient Descent (GD)

•	Basic optimizer.

•	Updates weights by computing the gradient of the loss function.

•	Can be slow and inefficient for large datasets.

Formula:

python

w = w - learning_rate * gradient

2. Stochastic Gradient Descent (SGD)

•	Instead of using the whole dataset, it updates weights using one sample at a time.

•	Faster than regular GD, but noisier.

Example:

python

optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

3. Mini-Batch Gradient Descent

•	A compromise between GD and SGD.

•	Updates weights using a small batch of data.

Example:

python

# This is the most commonly used approach in practice

4. Momentum

•	Improves SGD by adding a "momentum" term, which helps move faster in the right direction and reduces oscillation.

Example:

python

optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9)

5. Adagrad

•	Adapts learning rates for each parameter based on past gradients.

•	Good for sparse data (like NLP).

Example:

python

optimizer = torch.optim.Adagrad(model.parameters(), lr=0.01)

6. RMSprop

•	Similar to Adagrad but uses an exponentially decaying average of squared gradients.

•	Works well for recurrent neural networks (RNNs).

Example:

python

optimizer = torch.optim.RMSprop(model.parameters(), lr=0.001)

7. Adam (Adaptive Moment Estimation) ⭐ (Most popular)

•	Combines ideas from Momentum and RMSprop.

•	Automatically adjusts learning rate and has great performance in most deep learning tasks.

Example:

python

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)


Q. 19. What is sklearn.linear_model?

Answer :-

A module for implementing linear models like:
- LinearRegression
- LogisticRegression
- Ridge, Lasso, etc.

Q. 20. What does model.fit() do? What a argument must be given?

Answer :-

What does model.fit() do?

In libraries like Keras (TensorFlow), the model.fit() function is used to train your machine learning model.

 In simple terms:

model.fit() takes your input data and teaches the model how to make predictions by learning from the data.


What does it do exactly?

•	Feeds the training data to the model.

•	Performs forward propagation (makes predictions).

•	Calculates loss/error using a loss function.

•	Applies backpropagation using an optimizer to update weights.

•	Repeats for multiple epochs.

Required Arguments in model.fit()

Here are the minimum required arguments:

model.fit(x, y, ...)

Argument	Description

x	Input features (training data) — e.g. images, text, numbers

y	Target labels (what the model should learn to predict)

Common Optional Arguments:

Argument	Description

epochs	Number of times the model sees the full dataset (default is 1)
batch_size	Number of samples processed before updating weights
validation_data	Tuple of validation data (x_val, y_val) for checking accuracy on unseen data

verbose	Controls output during training (0 = silent, 1 = progress bar, 2 = one line per epoch)

callbacks	Functions like early stopping or saving checkpoints during training

Example:

model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_val, y_val))

This means:

•	Train for 10 epochs

•	Use batches of 32 samples

•	Evaluate performance on validation data during training

```
Q. 21. What does model.predict() do? What arguments must be given?

Answer :-

What does model.predict() do?

The function model.predict() is used to make predictions with a trained model.

In simple terms:

model.predict() takes new input data and returns the model’s predicted output.

What happens under the hood?

•	The model uses the weights it learned during training.
•	Applies a forward pass to the input data.
•	Returns the predicted result (e.g., class probabilities or numeric values).

Required Arguments for model.predict()

model.predict(x)

Argument	Description

x	The input data to make predictions on (e.g., images, text, numbers).

Must be in the same shape as what the model was trained on.

Optional Arguments (commonly used):

Argument	Description

batch_size	Number of samples processed at a time (default = 32)
verbose	Whether to print progress messages (0 = silent, 1 = progress bar)

What kind of output will you get?

•	Regression model → Single value (e.g., house price: 300,000)

•	Binary classification → Probability (e.g., [0.85] for class 1)

•	Multi-class classification → Array of probabilities (e.g., [0.1, 0.7, 0.2])

Q. 22. What are continuous and categorical variables?

Answer :-

Continuous Variables

•	These are quantitative variables that can take any value within a range.

•	They are usually measurable, not countable.

Categorical Variables

•	These are qualitative variables that represent categories or groups.

•	They are not numerical (though they can be encoded with numbers).

•	Two main types:

o	Nominal: No natural order (e.g., colors: red, blue, green).

o	Ordinal: Have a meaningful order (e.g., education level: high school < bachelor's < master's).
